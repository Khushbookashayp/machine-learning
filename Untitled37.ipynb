```python
#Question 10: You are working as a data scientist for a retail company. You receive a customer transaction dataset that contains
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.utils import resample

# Sample dataset
data = {
    "age": [25, np.nan, 35, 45, np.nan, 30, 50],
    "transaction_amount": [200, 5000, 300, 10000, 250, 400, 12000],
    "payment_method": ["Card", "UPI", "Card", "Cash", "UPI", "Card", "Cash"],
    "fraud": [0, 0, 0, 1, 0, 0, 1]
}

df = pd.DataFrame(data)

# ----- Handling Missing Ages -----
age_imputer = SimpleImputer(strategy="median")
df["age"] = age_imputer.fit_transform(df[["age"]])

# ----- Handling Outliers using IQR -----
Q1 = df["transaction_amount"].quantile(0.25)
Q3 = df["transaction_amount"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df["transaction_amount"] = np.clip(df["transaction_amount"], lower, upper)

# ----- Encoding Categorical Variables -----
df_encoded = pd.get_dummies(df, columns=["payment_method"], drop_first=True)

# ----- Handling Class Imbalance -----
majority = df_encoded[df_encoded.fraud == 0]
minority = df_encoded[df_encoded.fraud == 1]

minority_upsampled = resample(minority,
                              replace=True,
                              n_samples=len(majority),
                              random_state=42)

df_balanced = pd.concat([majority, minority_upsampled])

# ----- Feature Scaling -----
scaler = StandardScaler()
df_balanced[["age", "transaction_amount"]] = scaler.fit_transform(
    df_balanced[["age", "transaction_amount"]]
)

# ----- Train-Test Split -----
X = df_balanced.drop("fraud", axis=1)
y = df_balanced["fraud"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

print("Training data shape:", X_train.shape)
print("Testing data shape:", X_test.shape)
```
